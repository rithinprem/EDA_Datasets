```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/amazon.csv")
```

```python
#1. What is the average rating for each product category?


#need to convert rating dtype to int

df.rating = pd.to_numeric(df.rating,errors='coerce')
```

```python
df_avg_rating = pd.DataFrame(df.groupby('category')['rating'].mean()).reset_index()
df_avg_rating
```

```python
#2.  What are the top rating_count products by category?

def str_to_numbers(num):
  if isinstance(num,str):
    num=num.replace(",","")
  return float(num)


df.rating_count = df.rating_count.apply(str_to_numbers)

df.groupby('category').agg({'rating_count':'max','product_name':'first'}).reset_index()
```

```python
#3. What is the distribution of discounted prices vs actual prices ?

def convert(n):
  if isinstance(n,str):
    n = n.replace(",","").replace("₹","")
  return float(n)


df.discounted_price = df.discounted_price.apply(convert)
df.actual_price = df.actual_price.apply(convert)

plt.figure(figsize=(6, 6))
plt.scatter(df['actual_price'], df['discounted_price'], alpha=0.5,marker=".")
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.title('Distribution of Discounted Prices vs Actual Prices')
plt.grid(True)
plt.show()
```

```python
#4.  How does the average discount percentage vary across categories?

df.discount_percentage = df.discount_percentage.apply(lambda x: float(x.replace("%","") if isinstance(x,str) else x))

pd.DataFrame(df.groupby('category')['discount_percentage'].mean()).reset_index()
```

```python
#5. What are the most popular product names?



popular_products = df.groupby('product_name')['rating_count'].sum().reset_index()

popular_products = popular_products.sort_values(by='rating_count', ascending=False)

popular_products = pd.DataFrame(popular_products).reset_index().drop('index',axis=1)

popular_products[:10]   #top 10 famous products
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
#6.  What are the most popular product keywords?

import re

stop_words = set(stopwords.words('english'))


strings_dic={}

def split_words(string):
  l = string.split(" ")
  for word in l:
    word = re.sub(r'[^a-zA-Z0-9]', '', word)
    if word not in stop_words:
      if word not in strings_dic:
        strings_dic[word]=1
      else:
        strings_dic[word]+=1

  return string


df.product_name.apply(split_words)


#Most popular product keywords with their counts
list(dict(sorted(strings_dic.items(),key=lambda x:x[1],reverse=True)).items())[:10]
```

```python
#7.  What are the most popular product reviews ?

df.sort_values('rating_count',ascending=False)[['product_name','review_title']].drop_duplicates('review_title').reset_index().drop('index',axis=1)[:10]
```

```python
#8. What is the correlation between discounted_price and rating ?

print("Correlation:",df.discounted_price.corr(df.rating))

#positive and less correlated

sns.lineplot(x='rating',y='discounted_price',data=df).set_title('Discounted_price vs rating')
plt.show()
```

```python
#9. What are the Top 5 categories based on the highest ratings?

category_ratings = df.groupby('category')['rating'].mean()

top_categories = category_ratings.sort_values(ascending=False).head(5)

print(top_categories)
```

```python
#1.  Identify the top 5 popular artists based on the mean popularity of their tracks.Show the mean popularity of
#    tracks for the top 5 popular artists Using BarPlot.


df_popular_artists= pd.DataFrame(df1.groupby('artists')['popularity'].mean()).reset_index().sort_values('popularity',ascending=False)[:5]
df_popular_artists = df_popular_artists.reset_index().drop('index',axis=1)
sns.barplot(x='artists',y='popularity',data=df_popular_artists,palette='viridis')
plt.title('Top 5 Popular Artists')
plt.xticks(rotation=45)
plt.show()
```

```python
#2.  Determine the top 5 popular songs based on their popularity ratings.Display the popularity ratings of the top
#    5 popular songs using BarPlot.


df_popular_songs = pd.DataFrame(df1.groupby('track_name')['popularity'].mean()).reset_index()
df_popular_songs = df_popular_songs.sort_values('popularity',ascending=False)[:5]
sns.barplot(y='track_name',x='popularity',data=df_popular_songs,palette='viridis')
plt.title('Top 5 Popular Songs')
plt.show()
```

```python
#3. Find the top 5 trending genres based on the mean popularity of tracks within each genre.Visualize the mean
#    popularity of tracks for the top 5 trending genres.

df_genre = pd.DataFrame(df1.groupby('track_genre')['popularity'].mean()).reset_index()
df_genre=df_genre.sort_values('popularity',ascending=False)[:5]
sns.barplot(x='track_genre',y='popularity',data=df_genre,palette='viridis').set_title('Top 5 trending genres')
plt.show()
```

```python
#4  Identify the top 5 longest songs among the tracks of the top 5 popular artists.Represent the duration of the
 #  top 5 longest songs among the tracks of the top 5 popular artists using BarPlot

df_popular_artists=pd.DataFrame(df1.groupby('artists')['popularity'].mean().reset_index()).sort_values('popularity',ascending=False)[:5]
df_longest_songs = pd.DataFrame(df1[df1.artists.isin(df_popular_artists.artists.values)].groupby('artists')['duration_ms'].max()).reset_index()
df_longest_songs=df_longest_songs.sort_values('duration_ms',ascending=False)
sns.barplot(y='artists',x='duration_ms',data=df_longest_songs,palette='viridis').set_title('Top 5 longest songs among top 5 popular artists')
plt.xticks(rotation=90)
plt.show()
```

```python
#5  Determine the top 5 most danceable songs based on their danceability scores.Illustrate the danceability
 # scores of the top 5 most danceable songs using PieChart

df_danceable_songs = pd.DataFrame(df1.groupby('track_name')['danceability'].mean()).reset_index().sort_values('danceability',ascending=False)[:5]
plt.pie(df_danceable_songs['danceability'], labels=['{} ({:.2f})'.format(track, score) for track, score in zip(df_danceable_songs['track_name'],
df_danceable_songs['danceability'])], explode=[0.05, 0.05, 0.05, 0.05, 0.05], shadow=True, startangle=140)
plt.show()
```
