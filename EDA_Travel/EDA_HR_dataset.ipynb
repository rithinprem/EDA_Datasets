#3. Performance Analysis

Problem Statement: Evaluate the performance scores across different departments and positions, and examine if there's a correlation between performance and engagement survey scores.

```python
df.columns
```

```python
df.Department=df.Department.str.replace('Production       ','Production')
```

```python
df.Department.value_counts()
```

```python
df.PerformanceScore.unique()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Department',data=df,hue='PerformanceScore',palette='husl').set_title('Departmental Analysis')
plt.show()
```

```python
sns.scatterplot(x=df.PerformanceScore,y=df.EngagementSurvey).set_title('Survey scores vs Performance score')
plt.show()
```

```python
#top 10 employees having special projects

df[df['SpecialProjectsCount']!=0].sort_values('SpecialProjectsCount',ascending=False)[0:10]
```

```python
#compare top 10 highest and lowest salary
x = list(range(1,21))
Highest_Salary = df.sort_values('Salary',ascending=False)[0:10].assign(Category="Highest")
Lowest_Salary = df.sort_values('Salary',ascending=True)[0:10].assign(Category="Lowest")
combined = pd.concat([Highest_Salary,Lowest_Salary])
sns.scatterplot(x=x,y = 'Salary',data=combined,hue='Category')
plt.xticks(range(1, 21))
plt.xlabel('Rank')
plt.ylabel('Salary')
plt.title('Comparison of Top 10 Highest and Lowest Salaries')
plt.legend(title='Category')
plt.show()
```

#4. Turnover Analysis

Problem Statement: Identify patterns or factors associated with employee turnover, focusing on variables like department, employment status, and reasons for termination.

```python
df.columns
```

```python
df.DateofHire = pd.to_datetime(df.DateofHire)
df.DateofTermination = pd.to_datetime(df.DateofTermination)
df['EmploymentDuration'] = (df['DateofTermination'] - df['DateofHire']).dt.days
sns.lineplot(x='Department',y='EmploymentDuration',data=df,marker="o").set_title('Turnover Analysis')
plt.xticks(fontsize=7)
plt.show()
```

```python
sns.countplot(x='EmploymentStatus',data=df,hue='Department',palette='husl').set_title('Employment Status')
plt.ylabel('Number of Employees')
```

```python
df.EmpSatisfaction
```

```python
sns.countplot(x=df.EmpSatisfaction,hue=df.Department,palette='Set3').set_title('Departmental wise Rating')
plt.ylabel('No of Employees')
```
